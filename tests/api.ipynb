```python
from sqlalchemy import text, engine

from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String


#connect to the database

DATABASE_URL = "sqlite:///./globant_de.db"

engine = create_engine(DATABASE_URL)
metadata = MetaData()
connection = engine.connect()
```

```python
# lets insert a registry in jobs

jobs_table = Table(
    'jobs', metadata,
    Column('id', Integer, primary_key=True),
    Column('job', String),
    extend_existing=True
)

metadata.create_all(engine, checkfirst=True)

insert_stmt = jobs_table.insert().values(id=1, job='Software Engineer')

with engine.connect() as conn:
    try:
        conn.execute(insert_stmt)
    except Exception as e:
        pass
    conn.commit()
```

```python
with engine.connect() as conn:
    try:
        p = conn.execute(jobs_table.select().where(jobs_table.c.id == 1)).fetchone()
    except Exception as e:
        pass
    conn.commit()
print(p)  # should print (1, 'Software Engineer')
```
